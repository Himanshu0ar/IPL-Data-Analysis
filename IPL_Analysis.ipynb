# IPL Data Analysis Project

## Objective

The objective of this project is to analyze IPL match data and extract meaningful insights using Python, Pandas, NumPy, and Matplotlib.

## Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the Dataset

matches = pd.read_csv("data/matches.csv")
deliveries = pd.read_csv("data/deliveries.csv")

## Data Cleaning and Preprocessing

```python
matches.head()
```

```python
deliveries.head()
```

```python
matches.info()
```

```python
matches.isnull().sum()
```

## Most Successful Teams

```python
matches['winner'].value_counts()
```

```python
matches['winner'].value_counts().head(1)
```

```python
matches['winner'].value_counts().plot(kind='bar')

plt.title("Most Successful IPL Teams")
plt.xlabel("Teams")
plt.ylabel("Wins")

plt.xticks(rotation=90)
plt.show()
```

## Toss Winner Analysis

```python
matches['toss_winner'].value_counts()
```

```python
np.sum(matches['toss_winner'] == matches['winner'])
```

## Top Run Scorers

```python
top_runners = deliveries.groupby('batter')['batsman_runs'].sum()
topTen_runners = top_runners.sort_values(ascending=False).head(10)
```

```python
topTen_runners.plot(kind='bar')
plt.title("Top IPL Run Scorers")
plt.show()
```

## Top Sixes and Fours

```python
sixes = deliveries[deliveries['batsman_runs']==6]
sixes['batter'].value_counts().head(10)
```

```python
fours = deliveries[deliveries['batsman_runs']==4]
fours['batter'].value_counts().head(10)
```

```python
top_runner = top_runners.sort_values(ascending=False).head(1)
top_runner
```

## Venue Analysis

```python
matches['venue'].value_counts().head(10)
```

## Win Percentage

```python
(matches['winner'].value_counts()/matches['winner'].value_counts().sum())*100
```

```python
((matches['winner'].value_counts()/matches['winner'].value_counts().sum())*100).plot(kind='bar')
plt.title('Win Percentage Analysis')
plt.show()
```

## Season Wise Analysis

```python
matches['season'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,5))

matches['season'].value_counts().sort_index().plot(kind='line', marker='o')

plt.title("Matches Played Per IPL Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")

plt.grid(True)

plt.xticks(rotation=90)
plt.show()
```

### Insight:
The number of IPL matches has generally increased over the years.
